- Leer imagen a color con OpenCV

```python
import cv2
from google.colab.patches import cv2_imshow

imagen = cv2.imread('imagen_reto_colores.jpg', cv2.IMREAD_COLOR)

# cv2_imshow(imagen)
```

- Imprimir forma (shape)

```python
forma = imagen.shape

#print(forma)
```

- Extraer 5 zonas de interés homogeneas

```python
roi1 = imagen[870:920, 1405:1455] 
roi2 = imagen[640:690, 1530:1580]
roi3 = imagen[500:550, 1688:1738]
roi4 = imagen[160:210, 260:310]
roi5 = imagen[1152:1202, 2552:2602]

# cv2_imshow(roi1)
# cv2_imshow(roi2)
# cv2_imshow(roi3)
# cv2_imshow(roi4)
# cv2_imshow(roi5)
```

- Extraer 1 zonas de interes no homogenea

```python
roi6 = imagen[640:720, 1666:1746]

# cv2_imshow(roi6)
```

- Dibujr recuadro en las regiones en la imagen original

```python
img_copia = imagen.copy()

cv2.rectangle(img_copia, (1405, 870), (1455,920), (255,255,255), 2)
cv2.putText(img_copia, 'roi1', (1405,865), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.rectangle(img_copia, (1530, 640), (1580, 690), (255,255,255), 2)
cv2.putText(img_copia, 'roi2', (1530, 635), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.rectangle(img_copia, (1688, 500), (1738, 550), (255,255,255), 2)
cv2.putText(img_copia, 'roi3', (1688, 450), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.rectangle(img_copia, (260, 160), (310, 210), (255,255,255), 2)
cv2.putText(img_copia, 'roi4', (260, 155), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.rectangle(img_copia, (2552, 1152), (2602, 1202), (255,255,255), 2)
cv2.putText(img_copia, 'roi5', (2552, 1148), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.rectangle(img_copia, (1666, 640), (1746, 720), (255,255,255), 2)
cv2.putText(img_copia, 'roi6', (1666, 635), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

#cv2_imshow(img_copia)
```

- Obtener media y desviación estandar de cada uno de los canales en cada región

```python
rois = [roi1, roi2, roi3, roi4, roi5, roi6]

for i, roi in enumerate(rois, start=1):
  media, std = cv2.meanStdDev(roi)
  print(f"\nRoi {i}")
  print("Media", media.flatten())
  print("Desviación estándar", std.flatten())
```

### Analisis

- En el roi1 la media apunta a que es un color más que todo rojo, concordando con la imagen. Para el roi2 también hay concordancia pues las medias indicando un color grisaseo, asi como en el roi4 que es bastante oscuro y por ello tiene medias bastante bajas. E incluso también podemos decir que  hay concordancia en el roi3 porque tiene un poco más de intensidad en el canal rojo pero es muy oscuro como se puede ver en la imagen. Ahora, cuando vimos la imagen pensamos que el roi5 iba a tener una mucha predominancia en el canal verde, pero la verdad es que está muy igualado con el canal rojo, esto probablemente por el ruido que se ve.
- En cuanto a la desviación, se pudo observar que las roi1, 2, 3, y 4, tienen muy poca desviación, muestra una clara correspondencia entre los valores medios de los canales BGR y el color percibido visualmente en cada región, es decir, se comprueba que son homogeneas. Pero algo curioso pasa en la roi5 porque esta tiene una desviación bastante grande incluso más que la roi6, que si es claramente no homogenea por tener colores piel, grisaseos, cafés en la imagen.
