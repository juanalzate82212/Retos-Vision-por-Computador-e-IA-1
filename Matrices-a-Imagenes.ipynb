- Creación de Matriz 1000x1000 de numeros aleatorios de 0 a 255. Sin NumPy

```python
import random

def creacion_matriz(m, n):
    return [[creacion_random() for _ in range(n)] for _ in range(m)]

def creacion_random():
    x = random.randint(0,255)
    return x

def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

m = 1000
n = 1000
matriz = creacion_matriz(m, n)

#imprimir_matriz(matriz)

```

- Funciones para encontrar el minimo, máximo, media y desviación estandar de los valores de la matriz. Sin librerías.

```python
def encontrar_minimo(matriz, MIN):
    for fila in matriz:
        for numero in fila:
            if numero < MIN:
                MIN = numero
    return MIN

def encontrar_maximo(matriz, MAX):
    for fila in matriz:
        for numero in fila:
            if numero > MAX:
                MAX = numero
    return MAX

def encontrar_media(matriz):
    suma = 0
    cantidad = 0
    for fila in matriz:
        for numero in fila:
            suma += numero
            cantidad += 1
    return suma / cantidad

def encontrar_desviacion(matriz, MEDIA):
    suma_cuadrados = 0
    cantidad = 0
    for fila in matriz:
        for numero in fila:
            suma_cuadrados += (numero - MEDIA) ** 2
            cantidad += 1
    return (suma_cuadrados / cantidad) ** 0.5


MAX = 0
MIN = 255

MAX = encontrar_maximo(matriz, MAX)
MIN = encontrar_minimo(matriz, MIN)
MEDIA = encontrar_media(matriz)
DESVIACION = encontrar_desviacion(matriz, MEDIA)

print("El numero mayor es:", MAX)
print("El numero menor es:", MIN)
print("La media es:", MEDIA)
print("La desviación estándar es:", DESVIACION)

```
